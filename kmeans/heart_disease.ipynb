```python
X = df.iloc[:, :13]
y = df.iloc[:, 13]
X = X.replace(to_replace='?', value=0)
y = np.where(y > 0, 1, 0)
```

```python
pca = PCA(2)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
def plot_k_means(k):
    kmeans = KMeans(k)
    kmeans.fit(X)
    labels = kmeans.predict(X)
    X_2d = pca.fit_transform(X)
    plt.figure(figsize=(12.8,9.6))
    km = pd.DataFrame({'x':X_2d[:,0],'y':X_2d[:,1],'labels':labels,'true':y})
    sns.scatterplot(x='x', y='y', hue='labels',style='true',data=km, s=100)    
```

Going to try to first use 2.  As with the supervised learning there are two possible outcomes, heart disease or not.

```python
plot_k_means(2)
```

As can be seen, 2 does not perform well.  Going to try the inertia and elbow method to find the optimum number of clusters.  Just from the visual, there doesn't appear to be any seperate clusters.  

```python
# just using to visualize before kmeans
plot_k_means(1)
```

```python
results = np.empty((7,2))
for k in range(2,9):
    kmeans = KMeans(k)
    kmeans.fit(X)
    results[k-2,0]=k
    results[k-2,1] = kmeans.inertia_
sns.lineplot(x=results[:,0], y=results[:,1])
  
```
