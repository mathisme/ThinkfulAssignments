```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sqlalchemy import create_engine

import warnings
warnings.filterwarnings('ignore')
```

```python
postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'heartdisease'
```

```python
engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))
df = pd.read_sql_query('select * from heartdisease',con=engine)

engine.dispose()
```

```python
X = df.iloc[:, :13]
y = df.iloc[:, 13]
X = X.replace(to_replace='?', value=0)
y = np.where(y > 0, 1, 0)
```

```python
X = StandardScaler().fit_transform(X)
```

```python
pca = PCA(2)
```

```python
linkage_type = ['ward', 'complete', 'average']
```

```python
for method in linkage_type:
    plt.figure(figsize=(12,6))
    plt.title("Dendrogram using {}".format(method))
    dendrogram(linkage(X, method=method))
    plt.show()
```

I feel ward is the best, it breaks it into more evenly sized clusters.

```python
for method in linkage_type:
    agg = AgglomerativeClustering(linkage=method, 
                                      n_clusters=2)
    y_preds = agg.fit_predict(X)
    print('ARI for method',method,":",adjusted_rand_score(y,y_preds))
    print('Silhouette score for',method,":",silhouette_score(X,y_preds))
    
    
    
```

The ari score is best for ward but not as good as k means.  While the silhouette score for complete and average is better than the silhouette score for k means.
