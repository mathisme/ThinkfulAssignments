```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_openml
```

# Load the dataset and conduct any necessary preprocessing, such as normalizing the data.

```python
fmnist = fetch_openml('Fashion-MNIST', version=1, cache=True)
```

```python
fmnist.data.shape
```

```python
fmnist.target.unique()
```

```python
indices = np.random.choice(70000,10000, replace=False)
```

```python
X = (fmnist.data).iloc[indices]
y = (fmnist.target)[indices]
```

```python
X = StandardScaler().fit_transform(X)
```

# Apply PCA to the data and get the first two principal components.

```python
pca = PCA(n_components=2)
```

```python
X_pca = pca.fit_transform(X)
```

# Using the first two principal components, draw a graph of the data by coloring and labeling the data points as you did in the checkpoint.

```python
df = pd.DataFrame({
    'x':X_pca[:,0],
    'y':X_pca[:,1],
    'target':y
})
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='x',y='y',hue='target',data=df,s=100)
```

There does seem to be some mixing of target values so there is definite data loss.  Without the colors, classes would be indestinguishable.
