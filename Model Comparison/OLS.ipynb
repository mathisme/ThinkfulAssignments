```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data-original', header=None, 
            delim_whitespace=True, names=['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','car name']
)
```

```python
df.info()
```

# EDA

## Dealing with nulls

```python
df.isnull().sum()
```

Since we are using mpg as the dependant variable, I will drop the null values for mpg

```python
df.drop(df[df.mpg.isnull()].index,inplace=True)
```

```python
df.isnull().sum()
```

There are still six values where horsepower is null.

```python
df[df.horsepower.isnull()]
```

```python
df.horsepower.describe()
```

```python
# looked up the horsepower values online.  Now they may not be correct
df.loc[38,'horsepower'] = 100
df.loc[133,'horsepower'] = 94
df.loc[337,'horsepower'] = 53.5
df.loc[343,'horsepower'] = 88
df.loc[361,'horsepower'] = 81
df.loc[382,'horsepower'] = 82
```

```python
df.drop('car name',axis=1,inplace=True)
```

## Looking for outliers

```python
df.columns
```

```python
sns.boxplot(x="cylinders", y="mpg", data=df)
```

With respect to mpg there do appear to be outliers with the 4, 6, and 8 cylinder engine.

```python
plt.scatter('displacement','mpg',data=df)
```

There appears to be an inverse relationship or a logrithmic relationship between displacement and mpg.  I may want to explore this.

There are definite outliers where the displacement is around 250 and mpg is greater than 35 and also where the displacement is around 350 and the mpg is greater than 20.

```python
plt.scatter('horsepower','mpg',data=df)
```

Again the relationship looks like an inverse relationship or logrithmic relationship.  There is one obvious outlier where the horsepower is more than 125 and the mpg is more than 30.

```python
plt.scatter('weight','mpg',data=df)
```

The relationship is not quite linear, again it may be inverse or logrithmic.  There correlation is weaker than the variables explored before.  There may be some outliers where the weight is around 3000 to 3500 and the mpg is greater than 25.  With the weaker correlation it is harder to tell.

```python
plt.scatter('acceleration','mpg',data=df)
```

Very weak correlation. It's hard to judge outliers visually.
