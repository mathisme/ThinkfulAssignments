## Looking for outliers

```python
df.columns
```

```python
outliers = {}
```

```python
sns.boxplot(x="cylinders", y="mpg", data=df)
```

With respect to mpg there do appear to be outliers with the 4, 6, and 8 cylinder engine.

```python
for val in df.cylinders.unique():
    stats = df[df.cylinders==val]['mpg'].describe()
    q1 = stats['25%']
    q3 = stats['75%']
    iqr = 1.5 * (q3-q1)
    lst = list(df[(df.cylinders==val)&((df.mpg < q1 - iqr)|(df.mpg > q3 + iqr))].index)
    for i in lst:
        count, columns = outliers.get(i,(0,[]))
        count = count + 1
        columns.append('cylinders')
        outliers[i] = (count,columns)
outliers
```

```python
plt.scatter('displacement','mpg',data=df)
```

There appears to be an inverse relationship or a logrithmic relationship between displacement and mpg.  I may want to explore this.

There are definite outliers where the displacement is around 250 and mpg is greater than 35 and also where the displacement is around 350 and the mpg is greater than 20.

```python
lst = list(df[((df.displacement>225)&(df.mpg>35))|((df.displacement>325)&(df.mpg>20))].index)
for i in lst:
    count, columns = outliers.get(i,(0,[]))
    count = count + 1
    columns.append('displacement')
    outliers[i] = (count, columns)
outliers
```

```python
plt.scatter('horsepower','mpg',data=df)
```

Again the relationship looks like an inverse relationship or logrithmic relationship.  There is one obvious outlier where the horsepower is more than 125 and the mpg is more than 30.

```python
i = (df[(df.horsepower > 120)&(df.mpg > 30)].index)[0]
count, columns = outliers.get(i)
count = count + 1
columns.append('horsepower')
outliers[i] = (count,columns)
outliers
```

```python
plt.scatter('weight','mpg',data=df)
```

The relationship is not quite linear, again it may be inverse or logrithmic.  There correlation is weaker than the variables explored before.  There may be some outliers where the weight is around 3000 to 3500 and the mpg is greater than 25.  With the weaker correlation it is harder to tell.

```python
plt.scatter('acceleration','mpg',data=df)
```

Very weak correlation. It's hard to judge outliers visually.

```python
sns.boxplot(x="model year", y="mpg", data=df)
```

1978 has 1 outlier

```python
stats = df[df['model year']==78]['mpg'].describe()
q3 = stats['75%']
q1 = stats['25%']
iqr = 1.5*(q3 - q1)
i = (df[(df['model year']==78)&(df.mpg>q3+iqr)].index)[0]
count,columns = outliers.get(i,(0,[]))
count = count + 1
columns.append('model year')
outliers[i] = (count,columns)
outliers
```

As outliers were searched for with respect to the dependent variable, mpg in 4 columns, any outlier that appears in two columns will be dropped as a persistant outlier.

```python
for i in outliers.keys():
    count, _ = outliers.get(i)
    if count > 1:
        df.drop(i, inplace=True)
outliers
```

For OLS I will use weight_inv, model year, and acceleration

```python
df2 = df[['mpg','weight_inv','model year','acceleration']]
sns.heatmap(df2.corr(),square=True,annot=True)
```

# Model creation

```python
X = df2.drop('mpg',axis=1)
y = df2.mpg
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,random_state=625)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

# Model evaluation

```python
def mape(y_values, y_preds):
    return np.mean(np.abs(y_values-y_preds)/y_values)
```

```python
def print_prediction_errors(y_test, y_preds):
    print("The mean absolute error of the model: ",mean_absolute_error(y_test, y_preds))
    print("The mean absolute percentage error of the model: ",mape(y_test, y_preds))
    print("The mean square error of the model: ",mse(y_test,y_preds))
    print("The root mean square error of the model: ",rmse(y_test,y_preds))
```

```python
y_preds = lr.predict(X_test)
```

```python
r1 = lr.score(X_train,y_train)
r2 = lr.score(X_test,y_test)
print("R-squared for the training set is: ",r1)
print('-'*50)
print('R-squared for the test set is: ',r2)
print_prediction_errors(y_test,y_preds)
print('-'*50)
print("The difference in R squared between training and test sets is: ", r2-r1)
```
